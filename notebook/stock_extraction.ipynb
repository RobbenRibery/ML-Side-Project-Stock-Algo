```python
tickers = ['AAPL','FB','GOOGL','AMZN','MSFT']
start_date = '2014-1-1'
end_date = '2020-8-4'
data_total = web.DataReader(tickers, 'yahoo', start_date, end_date)
data_total.head()
```

```python
adj_close = data_total['Adj Close']
adj_close.head()
```

```python
adj_close.info()
```

```python
for col in adj_close.columns: 
    adj_close[col].plot(label = str(col),figsize=(16,8),title='Adjusted Close')
plt.legend()
```
